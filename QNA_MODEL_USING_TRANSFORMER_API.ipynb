```python
import gradio as gr
from transformers import pipeline

# Load QA pipeline
qa = pipeline(
    "question-answering",
    model="distilbert-base-uncased-distilled-squad"
)

def get_answer(context, question):
    if not context or not question:
        return "Please enter context and question"
    result = qa({
        "context": context,
        "question": question
    })
    return result["answer"]

# Gradio UI
gr.Interface(
    fn=get_answer,
    inputs=[
        gr.Textbox(lines=6, label="Context"),
        gr.Textbox(label="Question")
    ],
    outputs="text",
    title="Simple QnA App (Hugging Face)"
).launch()
```
